 1. Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("student_data.csv")
```

 2. Average Scores by Gender

```python
# Find average scores grouped by gender
avg_gender = data.groupby("gender")[["math score", "reading score", "writing score"]].mean()

# Plot as bar chart
avg_gender.plot(kind="bar", figsize=(8,5))
plt.title("Average Scores by Gender")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.legend(title="Subject")
plt.show()
```

 3. Scores vs Test Preparation

```python
# Find average scores based on test prep
avg_prep = data.groupby("test preparation course")[["math score", "reading score", "writing score"]].mean()

# Plot
avg_prep.plot(kind="bar", figsize=(8,5), color=["tomato", "orange", "skyblue"])
plt.title("Scores by Test Preparation Course")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.show()
```

 4. Box Plot: Scores by Parental Education

```python
# Change data shape for easy plotting
melted = pd.melt(data, id_vars="parental level of education",
                 value_vars=["math score", "reading score", "writing score"],
                 var_name="Subject", value_name="Score")

# Plot boxplot
plt.figure(figsize=(10,6))
sns.boxplot(x="parental level of education", y="Score", hue="Subject", data=melted)
plt.title("Scores by Parental Level of Education")
plt.xticks(rotation=45)
plt.show()
```

 5. Score Distribution by Lunch Type

```python
# Change shape again for lunch
melted_lunch = pd.melt(data, id_vars="lunch",
                       value_vars=["math score", "reading score", "writing score"],
                       var_name="Subject", value_name="Score")

# Violin plot
plt.figure(figsize=(8,6))
sns.violinplot(x="Subject", y="Score", hue="lunch", data=melted_lunch, split=True)
plt.title("Score Distribution by Lunch Type")
plt.show()
```

 6. Heatmap: Score Correlation

```python
# Check how math, reading, and writing scores are related
plt.figure(figsize=(5,4))
sns.heatmap(data[["math score", "reading score", "writing score"]].corr(),
            annot=True, cmap="YlGnBu")
plt.title("Score Correlation Heatmap")
plt.show()
```
